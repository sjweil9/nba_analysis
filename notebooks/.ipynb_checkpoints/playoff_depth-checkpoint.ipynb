```python
import pandas as pd
df = pd.read_csv('../data/nba_player_seasons.csv')
```

Drop all the rows where we don't have any playoffs for that year, as there is nothing to compare.

```python
df = df[df['PL_TOTAL_MINS'].notna()]
```

```python
df.head()
```

Add a calculated column for difference between playoffs and regular season

```python
df['RS_PL_NETRTG_DROPOFF'] = df['PL_NETRTG'] - df['RS_NETRTG']
```

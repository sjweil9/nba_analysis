```python
import pandas as pd

df = pd.read_csv("data/nba_lineups.csv")
```

```python
df.head()
```

```python
df.groupby('LINEUP_TYPE')['MIN'].sum().sort_values(ascending=False)
```

Remove lineup types where we have insufficient minutes.

```python
df.groupby('LINEUP_TYPE')['MIN'].sum().quantile(.25)
```

```python
df.groupby('LINEUP_TYPE').filter(lambda x: x['MIN'].sum() > 1000)
```

Group by lineup type and look at some basic averages and descriptive data

```python
df['AST_PER_48'] = (df['AST'] / df['MIN']) * 48
df.groupby('LINEUP_TYPE')['AST_PER_48'].mean().sort_values(ascending=False)
```

```python
df['PTS_PER_48'] = (df['PTS'] / df['MIN']) * 48
```

```python
df.groupby('LINEUP_TYPE')['PTS_PER_48'].mean().sort_values(ascending=False)
```
